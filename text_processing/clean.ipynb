```python
from mysql import connector
from text_preprocessing import clean_and_tokenize_text

# اسم الداتا سيت يلي بدك تنضف النصوص الخاصة فيها
dataset_name = "beir"

def clean_and_store_all():
    # اتصال بقاعدة البيانات
    conn = connector.connect(
        host="localhost",
        user="root",
        password="",
        database="ir"
    )
    cursor = conn.cursor(dictionary=True)

    # جلب كل الوثائق التابعة للداتا سيت المطلوب
    cursor.execute("SELECT id, text FROM documents WHERE dataset_name = %s", (dataset_name,))
    rows = cursor.fetchall()
    print(f"[i] عدد الوثائق التي سيتم تنظيفها: {len(rows)}")

    # تنظيف وتحديث كل وثيقة
    for idx, row in enumerate(rows, 1):
        doc_id = row['id']
        original_text = row['text']
        cleaned_tokens = clean_and_tokenize_text(original_text, dataset_name=dataset_name)
        cleaned_text = " ".join(cleaned_tokens)

        cursor.execute(
            "UPDATE documents SET process_text2 = %s WHERE id = %s",
            (cleaned_text, doc_id)
        )

        if idx % 100 == 0 or idx == len(rows):
            conn.commit()
            print(f"✅ تم تحديث {idx} / {len(rows)} وثيقة")

    # إغلاق الاتصال
    conn.commit()
    cursor.close()
    conn.close()
    print("🎉 تم تنظيف وتحديث جميع الوثائق بنجاح.")

# تشغيل الدالة
clean_and_store_all()
```
