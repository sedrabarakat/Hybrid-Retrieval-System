```python
import os
import sys
import joblib

project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
if project_root not in sys.path:
    sys.path.insert(0, project_root)

index_path = r"C:\Users\HP\IR-project\indexing\saved_models\inverted_index\beir_inverted_index.joblib"

if not os.path.exists(index_path):
    raise FileNotFoundError(f"لم يتم العثور على الفهرس المعكوس: {index_path}")

inverted_index = joblib.load(index_path)

print(f"عدد المصطلحات في الفهرس: {len(inverted_index)}")

# جمع جميع أنواع الـ id من أول 1000 مصطلح (أو أقل إذا الفهرس أصغر)
id_types = set()
sample_count = 0
max_terms_to_check = 1000

for term, doc_ids in inverted_index.items():
    for doc_id in doc_ids:
        id_types.add(type(doc_id))
    sample_count += 1
    if sample_count >= max_terms_to_check:
        break

print(f"أنواع الـ id المستخدمة في الفهرس (من عينة حتى {max_terms_to_check} مصطلح):")
for t in id_types:
    print(f"- {t}")
```
