```python
import os
import numpy as np
import joblib

# حدد المجلد يلي فيه ملفات .npy
folder = "saved_models/embedding"

# تأكد إنو المجلد موجود
if not os.path.exists(folder):
    print(f"[!] المجلد غير موجود: {folder}")
else:
    for file in os.listdir(folder):
        if file.endswith(".npy"):
            full_npy_path = os.path.join(folder, file)
            
            # تحميل البيانات من ملف npy
            data = np.load(full_npy_path)

            # إنشاء اسم جديد بلاحقة .joblib
            joblib_file = file.replace(".npy", ".joblib")
            full_joblib_path = os.path.join(folder, joblib_file)

            # تخزين البيانات بصيغة joblib
            joblib.dump(data, full_joblib_path, compress=3)

            print(f"[✓] تم تحويل: {file} → {joblib_file}")
```
